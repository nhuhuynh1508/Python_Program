```python
import numpy as np
import matplotlib.pyplot as plt

def square(x_data):
  y_data = []
  for value in x_data:
    y_data.append(value*value) #value**2

  return y_data

x_data = list(np.arange(-10, 10, 0.5))
y_data = square(x_data)

# show data
plt.plot(x_data, y_data)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def exponential(x_data):
  y_data = []
  for value in x_data:
    y_data.append(math.exp(value))

  return y_data

x_data = list(np.arange(-10, 10, 0.5))
y_data = exponential(x_data)

# show data
plt.plot(x_data, y_data)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def sigmoid(x_data):
  y_data = []
  for value in x_data:
    sigmoid_value = 1.0 / (1 + math.exp(-value))
    y_data.append(sigmoid_value)

  return y_data

def derivative(y_data):
  derivative_data = []
  for value in y_data:
    derivative_value = value * (1 - value)
    derivative_data.append(derivative_value)

  return derivative_data

x_data = list(np.arange(-10, 10, 0.1))
y_data = sigmoid(x_data)
derivative_data = derivative(y_data)

# show data
plt.plot(x_data, y_data)
plt.plot(x_data, derivative_data)
plt.show()

```

```python
# Gradient computation x->y->z
import tensorflow as tf

# data
x = tf.Variable(2.0)

# construct function
with tf.GradientTape() as tape:
  y = x*x
  z = y*y

# compute derivative
dz_dx = tape.gradient(z, x)

# print out
print(dz_dx.numpy())
```

```python
# Gradient computation
import tensorflow as tf
import math

# data
x = tf.Variable(1.0)

# construct function
with tf.GradientTape() as tape:
  g = x**2 + 1 #x*x + 1
  h = tf.math.exp(g) #tf.exp(g)

# comput derivative
dh_dx = tape.gradient(h, x)

# print
print(dh_dx.numpy())
```

```python
# derivative
def square_function(x):
  return x*x

def compute_derivative(epsilon, x):
  # compute f(x)
  f_x = square_function(x)

  # compute f(x + epsilon)
  f_xpe = square_function(x + epsilon)

  derivative = (f_xpe - f_x) / epsilon
  return derivative

# test
x = 3 # y = 6
derivative = compute_derivative(0.0001, x)
print(derivative)
```

```python
# derivative function
def square_function(x):
  return x*x*x - 2*x + math.exp(x)

def compute_derivative(epsilon, x):
  # compute f(x)
  f_x = square_function(x)

  # compute f(x + epsilon)
  f_xpe = square_function(x + epsilon)

  derivative = (f_xpe - f_x) / epsilon
  return derivative

# test
x = 3 # y = 6
derivative = compute_derivative(0.0001, x)
print(derivative)
```
